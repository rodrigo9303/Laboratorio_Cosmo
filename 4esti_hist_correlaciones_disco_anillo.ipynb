```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=1000
ale_x=np.random.uniform(0.,10.,N)
ale_y=np.random.uniform(0.,10.,N)
plt.plot(ale_x,ale_y,'.')
```

```python
circle_r = 25
circle_x = 5
circle_y = 5

#N = 1000

phi = 2 * np.pi * np.random.rand(N)
r = circle_r * np.random.rand(N)
```

```python
x = np.sqrt(r) * np.cos(phi) + circle_x
y = np.sqrt(r) * np.sin(phi) + circle_y
```

```python
plt.plot(x, y, '.')
plt.xlabel(r"x", fontsize = 15)
plt.ylabel(r"y", fontsize = 15)
#plt.xlim(-1.5, 1.5)
#plt.ylim(-1.5, 1.5)
```

```python
plt.plot(ale_x,ale_y,'.')
plt.plot(x,y,'.')
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dtot = N*(N-1)/2
j = 0
dis_datos = []
dis_aleatorios = []
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(x[i],x[j],y[i],y[j])
        d_a = distancia(ale_x[i],ale_x[j],ale_y[i],ale_y[j])
        dis_datos.append(d_d)
        dis_aleatorios.append(d_a)
```

```python
data_n, data_bin,m = plt.hist(dis_datos,bins=20)
```

```python
ran_n, ran_bin, p = plt.hist(dis_aleatorios,bins=20)
```

```python
plt.plot(data_bin[1:],(data_n/ran_n) -1, label=('Peebles=$\dfrac{DD(r)}{RR(r)}-1$'))
plt.legend()
```

```python
dis_da = []
```

```python
for i in range(N):
    for j in range(N):
        d_dr = distancia(x[i],ale_x[j],y[i],ale_y[j])
        dis_da.append(d_dr)
```

```python
da_n, da_bin, q = plt.hist(dis_da,bins=20)
```

```python
plt.plot(data_bin[1:],(data_n*ran_n)/(da_n)**2,label=('Hamilton=$\dfrac{DD(r)RR(r)}{DR(r)^2}$'))
plt.legend()
```

```python
plt.plot(data_bin[1:],1.+ data_n/ran_n -2*da_n/ran_n, label=('Landy-Szalay=$1+\dfrac{DD(r)}{RR(r)}-2\dfrac{DR(r)}{RR(r)}$'))
plt.legend()
```

```python
plt.plot(data_bin[1:],((data_n)/(da_n))-1,label=('Davis-Peebles=$\dfrac{DD(r)}{DR(r)}-1$'))

plt.legend()
```
