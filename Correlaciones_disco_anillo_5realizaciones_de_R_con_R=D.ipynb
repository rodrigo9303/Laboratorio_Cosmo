```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=1000
ran_x= []
ran_y=[]
for i in range (5):
    ale_x=np.random.uniform(0.,10.,N)
    ale_y=np.random.uniform(0.,10.,N)
    ran_x.append(ale_x)
    ran_y.append(ale_y)
#print('para x =',ran_x)
#print('para y =',ran_y)
#plt.plot(ale_x,ale_y,'.')
```

```python
#print('para x =',ran_x[3])
#print('para y =',ran_y[3])
```

```python
plt.plot(ran_x[0],ran_y[0],'.')
plt.plot(ran_x[1],ran_y[1],'.')
plt.plot(ran_x[2],ran_y[2],'.')
plt.plot(ran_x[3],ran_y[3],'.')
plt.plot(ran_x[4],ran_y[4],'.')
```

```python
circle_r = 9
circle_x = 5
circle_y = 5

phi = 2 * np.pi * np.random.rand(N)
r = circle_r * np.random.rand(N)

x = np.sqrt(r) * np.cos(phi) + circle_x
y = np.sqrt(r) * np.sin(phi) + circle_y

plt.plot(x, y, '.')
plt.xlabel(r"x", fontsize = 15)
plt.ylabel(r"y", fontsize = 15)
```

```python
plt.plot(ran_x[0],ran_y[0],'.')
plt.plot(ran_x[1],ran_y[1],'.')
plt.plot(ran_x[2],ran_y[2],'.')
plt.plot(ran_x[3],ran_y[3],'.')
plt.plot(ran_x[4],ran_y[4],'.')
plt.plot(x,y,'.',color='k')
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dis_datos = []
dis_aleatorios = []
uno = ran_x[0]
dos = ran_y[0]
tres = ran_x[1]
cuatro = ran_y[1]
cinco = ran_x[2]
seis = ran_y[2]
siete = ran_x[3]
ocho = ran_y[3]
nueve = ran_x[4]
diez = ran_y[4]
for i in range(N):
    for j in range(i+1,N):
        #d_d = distancia(x[i],x[j],y[i],y[j])
        d_a1 = distancia(uno[i],uno[j],dos[i],dos[j])
        d_a2 = distancia(tres[i],tres[j],cuatro[i],cuatro[j])
        d_a3 = distancia(cinco[i],cinco[j],seis[i],seis[j])
        d_a4 = distancia(siete[i],siete[j],ocho[i],ocho[j])
        d_a5 = distancia(nueve[i],nueve[j],diez[i],diez[j])
        #dis_datos.append(d_d)
        dis_aleatorios.append(d_a1)
        dis_aleatorios.append(d_a2)
        dis_aleatorios.append(d_a3)
        dis_aleatorios.append(d_a4)
        dis_aleatorios.append(d_a5)
#print(ale_x)
#print(uno)
#print(dis_aleatorios)
print(len(dis_aleatorios))
```

```python
nbins = np.linspace(0.,np.sqrt(2)*10, 20)
ran_n, ran_bin,m = plt.hist(dis_aleatorios,nbins, label='$RR(r)$')
print(nbins)
plt.legend()
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(x[i],x[j],y[i],y[j])
        dis_datos.append(d_d)
```

```python
data_n, data_bin,m = plt.hist(dis_datos,nbins, label='$DD(r)$')
print(nbins)
plt.legend()
```

```python
dis_da = []
for i in range(N):
    for j in range(N):
        d_dr1 = distancia(x[i],uno[j],y[i],dos[j])
        d_dr2 = distancia(x[i],tres[j],y[i],cuatro[j])
        d_dr3 = distancia(x[i],cinco[j],y[i],seis[j])
        d_dr4 = distancia(x[i],siete[j],y[i],ocho[j])
        d_dr5 = distancia(x[i],nueve[j],y[i],diez[j])
        dis_da.append(d_dr1)
        dis_da.append(d_dr2)
        dis_da.append(d_dr3)
        dis_da.append(d_dr4)
        dis_da.append(d_dr5)
```

```python
da_n, da_bin, q = plt.hist(dis_da,nbins, label='$DR(r)$')
print(nbins)
plt.legend()
```

```python
Nest= ((N*(N-1))/(5*N*(N-1)))
plt.plot(data_bin[1:],Nest*(data_n/ran_n) -1, label=('Peebles=$\dfrac{1}{N_{est}^{2}}\dfrac{DD(r)}{RR(r)}-1$'))
plt.legend()
```

```python
plt.plot(data_bin[1:],(data_n*ran_n)/(da_n)**2,label=('Hamilton=$\dfrac{DD(r)RR(r)}{DR(r)^2}$'))
plt.legend()
```

```python
plt.plot(data_bin[1:],1.+ Nest*data_n/ran_n -2*Nest*da_n/ran_n, label=('Landy-Szalay=$1+\dfrac{1}{N_{est}^{2}}\dfrac{DD(r)}{RR(r)}-2\dfrac{1}{N_{est}^{2}}\dfrac{DR(r)}{RR(r)}$'))
plt.legend()
```

```python
plt.plot(data_bin[1:],Nest*((data_n)/(da_n))-1,label=('Davis-Peebles=$\dfrac{1}{N_{est}^{2}}\dfrac{DD(r)}{DR(r)}-1$'))
plt.legend()
```

```python
ranillo = 2*np.pi*np.random.uniform(1.1,1.4,N)
rani = np.sqrt(ranillo)#**(0.5)
w = rani * np.cos(phi) + circle_x
z = rani * np.sin(phi) + circle_y
plt.plot(w,z,'.')
plt.xlabel('x')
plt.ylabel('y')
```
