```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.linspace(0, np.pi, 500) #tiempo
f = 7.0 #frecuencia
A = 4.0 #amplitud
s = A*np.sin(2 * np.pi * f * t)#Función que será tranformada.

plt.ylabel("Amplitud $f(t)$")
plt.xlabel("Tiempo $t$")
plt.plot(t, s)
```

```python
Y = np.fft.fft(s)
plt.plot(Y, label='Transformda')
plt.xlabel('$\omega$')
plt.ylabel('$F(\omega)$')
plt.legend()
```

```python
s1 = A*np.cos(2 * np.pi * f * t)
plt.ylabel("Amplitud $f(t)$")
plt.xlabel("Tiempo $t$")
plt.plot(t, s1)
```

```python
Y = np.fft.fft(s1)
plt.plot(Y, label='Transformda')
plt.xlabel('$\omega$')
plt.ylabel('$F(\omega)$')
plt.legend()
```

```python
x = np.zeros(500)
x[100:150] = 1
x[200:260] = 2

plt.plot(x)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
X = np.fft.fft(x)

plt.plot(np.fft.fftshift(np.abs(X)))
plt.xlabel('k')
plt.ylabel('F(k)')
```
