```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=2000
L=100
datos=L*np.random.rand(N,2)
#x=np.linspace(0.,N,11)
plt.plot(datos,'.')
```

```python
aleatorios = L*np.random.rand(N,2)
```

```python
def distancia(x1,x2,y1,y2):
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
dtot = N*(N-1)/2
j = 0
dold = 0
dmax = 0
dis_datos = []
dis_aleatorios = []
```

```python
for i in range(N):
    for j in range(i+1,N):
        d_d = distancia(datos[i][0],datos[j][0],datos[i][1],datos[j][1])
        d_a = distancia(aleatorios[i][0],aleatorios[j][0],aleatorios[i][1],aleatorios[j][1])
        dis_datos.append(d_d)
        dis_aleatorios.append(d_a)
```

```python
data_n, data_bin,m = plt.hist(dis_datos,bins=20)
```

```python
ran_n, ran_bin, p = plt.hist(dis_aleatorios,bins=20)
```

```python
plt.plot(data_bin[1:],(data_n/ran_n) -1, label=('Peebles=$\dfrac{DD(r)}{RR(r)}-1$'))
plt.legend()
```

```python
dis_datosale = []
```

```python
for i in range(N):
    for j in range(N):
        d_datosale = distancia(datos[i][0],aleatorios[j][0],datos[i][1],aleatorios[j][1])
        dis_datosale.append(d_datosale)
```

```python
dr_n, dr_bin, q = plt.hist(dis_datosale,bins=20)
```

```python
plt.plot(data_bin[1:],(data_n*ran_n)/(dr_n)**2,label=('Hamilton=$\dfrac{DD(r)RR(r)}{DR(r)^2}$'))
plt.legend()
```

```python
plt.plot(data_bin[1:],1.+ data_n/ran_n -dr_n/ran_n, label=('Landy-Szalay=$1+\dfrac{DD(r)}{RR(r)}-\dfrac{DR(r)}{RR(r)}$'))
plt.legend()
```
