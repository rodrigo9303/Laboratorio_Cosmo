```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def transformada(x):#x es un arreglo de dimensi√≥n 1
    N = x.size
    n = np.arange(N)
    w = n.reshape((N,1))
    exponencial = np.exp(-2j*np.pi*w*n/N)
    result = np.dot(exponencial,x)
    return result
```

```python
t = np.linspace(-4*np.pi,4*np.pi,500)
r = np.sin(t)
```

```python
plt.plot(t,r)
```

```python
F = transformada(r)
```

```python
plt.plot(F)
```

```python
def transformada_inversa(x):#x es un arreglo de dimensi√≥n 1
    N = x.size
    n = np.arange(N)
    w = n.reshape((N,1))
    exponencial = np.exp(2j*np.pi*w*n/N)/N
    result = np.dot(exponencial,x)
    return result
```

```python
plt.plot(F)
```

```python
I=transformada_inversa(transformada(r))
```

```python
plt.plot(I)
```
