```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def transformada(x):#x es un arreglo de dimensi√≥n 1
    N = x.size
    n = np.arange(N)
    w = n.reshape((N,1))
    exponencial = np.exp(-2j*np.pi*w*n/N)
    result = np.dot(exponencial,x)
    return result
```

```python
t = np.linspace(-4*np.pi,4*np.pi,500)
r = np.exp(-t**2)
plt.plot(t,r,label=('$exp(-x^2)$'))
plt.xlabel('x')
plt.ylabel('y(x)')
plt.legend()
```

```python
T= transformada(r)
plt.plot(T,label=('Transformada'))
plt.xlabel('k')
plt.ylabel('F(k)')
plt.legend()
```

```python
t = np.linspace(-4*np.pi, 4*np.pi, 500) #tiempo
s = np.exp(-t**2)#Funci√≥n que ser√° tranformada.

plt.ylabel("$f(x)$")
plt.xlabel("$x$")
plt.plot(t, s)
```

```python
Y = np.fft.fft(s)
plt.plot(Y, label='Transformda')
plt.xlabel('$k$')
plt.ylabel('$F(k)$')
plt.legend()
```
